```python
import random
from tqdm import tqdm
```

```python
g = DirectedGraph()
for i in tqdm(range(20_000), total=len(range(20_000)) ):
    g.add_node(identifier=i, age=random.randint(17, 30), sex=random.choice([True, False]))
for i in tqdm(range(20_000), total=len(range(20_000)) ):
    nl = random.choice(list(g.nodes.keys()))
    nr = random.choice(list(g.nodes.keys()))
    for j in range(random.randint(1, 3)):
        g.add_edge(
            node_l=nl, node_r=nr, identifier=i+j,
            weight=random.randint(100, 500)
            )
# g.add_node(identifier='Kate', name='Kate', age=19, sex=False)
# g.add_node(identifier='Pedro', name='Pedro', age=21, sex=True)
# g.add_node(identifier='Gennadiy', name='Gennadiy', age=20, sex=True)
# g.add_node(identifier='Anastasia', name='Anastasia', age=18, sex=False)
```
