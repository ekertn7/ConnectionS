```python
from abc import ABC, abstractmethod


class NodesDescriptor:
    def __set_name__(self, owner, name):
        self._name = f'_{name}'

    def __get__(self, instance, owner):
        return getattr(instance, self._name)

    def __set__(self, instance, value):
        setattr(instance, self._name, [v*v for v in value])

    def __delete__(self, instance) -> None:
        raise Exception()

class DirectedEdgesDescriptor:
    @staticmethod
    def validate(value, nodes):
        if value in nodes:
            return True

    def __set_name__(self, owner, name):
        self._name = f'_{name}'

    def __get__(self, instance, owner):
        return getattr(instance, self._name)

    def __set__(self, instance, value):
        setattr(instance, self._name, self.validate(value, getattr(instance, '_nodes')))

    def __delete__(self, instance) -> None:
        raise Exception()

class Graph(ABC):
    nodes = NodesDescriptor()

    def __new__(cls, *args, **kwargs):
        return super().__new__(cls)

    def __init__(self, nodes, edges):
        self.nodes = nodes
        self.edges = edges


class DirectedGraph(Graph):
    edges = DirectedEdgesDescriptor()

    def __init__(self, nodes, edges):
        super().__init__(nodes, edges)
```
